```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
H_device, S_device = np.load("toy_models/strained_tests/dft/unstrained_device/output/no_lowdin/hs_los_no_lowdin.npy", allow_pickle=True)
active_index = np.load("toy_models/strained_tests/dft/unstrained_device/output/no_lowdin/index_active_region.npy")

# Extract active region: slice both rows and columns using active_index
H_active = H_device[0][np.ix_(active_index, active_index)]
S_active = S_device[0][np.ix_(active_index, active_index)]

# Plot
plt.imshow(H_active.real)
plt.colorbar()
plt.title("H_device (Active Region)")
plt.show()

plt.imshow(S_active.real)
plt.colorbar()
plt.title("S_device (Active Region)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

# Load data
H_device, S_device = np.load("toy_models/strained_tests/dft/unstrained_device/output/no_lowdin/hs_los_no_lowdin.npy", allow_pickle=True)
active_index = np.load("toy_models/strained_tests/dft/unstrained_device/output/no_lowdin/index_active_region.npy")

vmin = 1e-3
# Extract active region
H_active = H_device[0][np.ix_(active_index, active_index)]
S_active = S_device[0][np.ix_(active_index, active_index)]

# Plot H_device on log scale
plt.imshow(np.abs(H_active)+1e-12, norm=colors.LogNorm(vmin=vmin))
plt.colorbar()
plt.title("H_device (Active Region, Log Scale |H|)")
plt.show()

# Plot S_device on log scale
plt.imshow(np.abs(S_active)+1e-12, norm=colors.LogNorm(vmin=vmin))
plt.colorbar()
plt.title("S_device (Active Region, Log Scale |S|)")
plt.show()
```
