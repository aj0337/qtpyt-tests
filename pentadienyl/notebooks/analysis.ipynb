```python
index_active_region = np.load("./output/lowdin/index_active_region.npy")
de = 0.01
energies = np.arange(-15, 15 + de / 2.0, de).round(7)

index_C_2pz = np.array([1147, 1160, 1183, 1196, 1219])
index_N_2pz = np.array([1119,1232])
index_molecule = np.arange(1116, 1235)

n_orbitals = 2368
all_indices = np.arange(n_orbitals)
index_Au = np.setdiff1d(all_indices, index_molecule)
```

```python
pdos = np.load("./output/lowdin/pdos_spectral.npy")

# Sum PDOS over the active region
pdos_active = pdos[:, index_active_region]  # shape (601, 7)
pdos_active_sum = np.sum(pdos_active, axis=1)  # shape (601,)


pdos_N_2pz = pdos[:, index_N_2pz]
pdos_N_2pz_sum = np.sum(np.abs(pdos_N_2pz), axis=1)

pdos_C_2pz = pdos[:, index_C_2pz]
pdos_C_2pz_sum = np.sum(np.abs(pdos_C_2pz), axis=1)

pdos_molecule = pdos[:, index_molecule]
pdos_molecule_sum = np.sum(np.abs(pdos_molecule), axis=1)

pdos_Au = pdos[:, index_Au]
pdos_Au_sum = np.sum(np.abs(pdos_Au), axis=1)

dos = np.sum(np.abs(pdos), axis=1)

plt.figure(figsize=(6, 4))
plt.plot(energies, np.abs(pdos_N_2pz_sum),label='N_2pz')
plt.plot(energies, np.abs(pdos_C_2pz_sum),label='C_2pz')
plt.plot(energies, np.abs(pdos_molecule_sum),label='Molecule')
# plt.plot(energies, np.abs(pdos_Au_sum),label='Au')
# plt.plot(energies, np.abs(dos),label='Total DOS')
plt.legend()
plt.xlabel("Energy (eV)")
plt.ylabel("PDOS (N_2pz region)")
plt.title("Projected DOS (Summed over N_2pz Region)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sigma_ed = np.load("./output/lowdin/beta_1000/ed/sigma_ed.npy")
trace_sigma_ed = np.trace(sigma_ed,axis1=1, axis2=2)
energies = np.arange(-20, 20.1, 0.1)
plt.plot(energies,trace_sigma_ed.real,label='Real part')
plt.plot(energies,trace_sigma_ed.imag,label='Imaginary part')
plt.legend()
plt.xlim(-20,-18)
plt.ylim(-50,50)
plt.xlabel("Energy (eV)")
plt.ylabel("Trace of Sigma_ed")
plt.title("Trace of Sigma_ed")
plt.grid(True)
plt.show()
```
