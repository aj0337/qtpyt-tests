```python
output_folder = "output/compute_run/toy_model5"
# Define file paths
hyb_filename = os.path.join(output_folder, 'hybridization.bin')
mat_hyb_filename = os.path.join(output_folder, 'matsubara_hybridization.bin')

energies = np.arange(-10, 10, 0.01)
n_energies = len(energies)  # Number of energies
n_matsenergies = 3000
matrix_size = 9  # Size of the matrices (n_A x n_A)

# Read Hybridization data
with open(hyb_filename, 'rb') as f:
    HB_retarded = np.fromfile(f, dtype=np.complex128).reshape((n_energies, matrix_size, matrix_size))

# Read Matsubara Hybridization data
with open(mat_hyb_filename, 'rb') as f:
    HB_matsubara = np.fromfile(f, dtype=np.complex128).reshape((n_matsenergies, matrix_size, matrix_size))
```

```python
HB_retarded.shape
```

```python
HB_matsubara.shape
```

```python
import matplotlib.pyplot as plt

# Compute the trace of the matrices for both retarded and Matsubara hybridizations
trace_retarded_real = np.real(np.trace(HB_retarded, axis1=1, axis2=2))
trace_retarded_imag = np.imag(np.trace(HB_retarded, axis1=1, axis2=2))

trace_matsubara_real = np.real(np.trace(HB_matsubara, axis1=1, axis2=2))
trace_matsubara_imag = np.imag(np.trace(HB_matsubara, axis1=1, axis2=2))

# Plot the trace of the retarded hybridization
plt.figure(figsize=(10, 6))
plt.plot(energies, trace_retarded_real, label='Real', linewidth=1.5)
plt.plot(energies, trace_retarded_imag, label='Imaginary', linewidth=1.5, linestyle='--')
plt.xlabel("Energy (eV)")
plt.ylabel("Trace")
plt.title("Trace of Retarded Hybridization")
plt.legend()
plt.grid()
plt.show()

# For Matsubara, plot against Matsubara index (since these are not energies)
matsubara_indices = np.arange(n_matsenergies)

plt.figure(figsize=(10, 6))
plt.plot(matsubara_indices, trace_matsubara_real, label='Real', linewidth=1.5)
plt.plot(matsubara_indices, trace_matsubara_imag, label='Imaginary', linewidth=1.5, linestyle='--')
plt.xlabel("Matsubara Index")
plt.ylabel("Trace (Matsubara)")
plt.title("Trace of Matsubara Hybridization")
plt.legend()
plt.grid()
plt.show()
```
