```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
smearing = [ 0.07]
Efermi  = 4.344856011792648
plt.figure(figsize=(8,4))
for s in smearing:
    directory = f'smearing_{s}'
    energy,bridge_pdos = np.load(f"{directory}/bridge_pdos.npy")
    energy -= Efermi
    plt.plot(energy,bridge_pdos,label=f"smearing = {s}")
plt.xlim(-2,2)
plt.ylim(0,2.5)
plt.xlabel("E - Ef(eV)")
plt.ylabel("1,3 BDA DOS")
plt.legend()
plt.show()
```
