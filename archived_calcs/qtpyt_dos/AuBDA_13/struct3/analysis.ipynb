```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
reference = np.loadtxt("transmission_function/lcao/ET_hs.dat")
```

```python
etas = [1e-3]
plt.figure(figsize=(8,4))
plt.plot(reference[:,0],reference[:,1],label='Reference')
for eta in etas:
    E,T = np.load(f'transmission_function/lcao/ET_eta{eta}_lcao.npy')
    plt.plot(E,T,label=rf'$\eta = {eta}$')
    if any(value > 1 for value in T):
        print(f"Warning: Values in T for eta={eta} exceed 1.")

plt.xlim(-3,3)
plt.title("LCAO")
plt.xlabel(r'$E-E_{f}(eV)$')
plt.ylabel('T(E)')
plt.yscale('log')
plt.ylim(5e-6,1)
plt.legend()
```
