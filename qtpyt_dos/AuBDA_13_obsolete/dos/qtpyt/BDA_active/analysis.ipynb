```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))

dos_BDA = np.load("pdos_eta_0.025.npy")
dos1_BDA = np.load("pdos_eta_0.1.npy")
energies = np.load("energies.npy")
plt.plot(energies,dos_BDA,label="1,3 BDA smearing 0.025")
plt.plot(energies,dos1_BDA,label="1,3 BDA smearing 0.1")
plt.legend()
plt.ylim(0,2.5)
plt.xlim(-2,2)
plt.xlabel(r"E-$E_{F}$(eV)")
plt.ylabel("DOS")
```
